### 1: magic arguments

There are many different "magic" commands we can pass to matplotlib using the % sign at the beginning of our code. Below are the two most common examples.

````python
import matplotlib.pyplot as plt
import random
xData = [random.randint(1,10) for i in range(100)] #make 100 random x points
yData = [random.randint(1,10) for i in range(100)] #make 100 random y points
````

````python
%matplotlib inline #this one lets our figures show up nicely in the notebooks without explicitly calling plt.show()
plt.plot(xData,yData) 
````

````python
%matplotlib notebook #this returns an interactive figure, but it always stays below this cell
plt.plot(xData,yData)
````

````python
plt.scatter([1],[1], color = "red") #notice this added a point to our plot above
#the plot doesn't redisplay below
````

````python
plt.show() #notice if we call plt.show it is still up above
````

### 2: all the different kinds of plots

The following examples have been taken and lightly modified from the matplotlib guide.

````python
#making a linear and semilog plot
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# close all the figures, if open from previous commands
plt.close('all') 

x = np.linspace(0.01, 5, 100) #generate 100 values in a range from 0.01 to 5
e = np.exp(x) #exponential function

#linear plot
plt.plot(x,e)
plt.xlabel("x")
plt.ylabel("y=exp(x)")
plt.title("Linear Y axis")

#semilog plot
#log(exp(x))=x therefore straight line will be displayed
plt.figure()
plt.semilogy(x,e) #semilogy: semilog y-axis
plt.xlabel("x")
plt.ylabel("y=exp(x)") 
plt.title("Log Y axis")
             
plt.show() #display the plot
````

````python
#making a histogram
plt.close("all")

fig=plt.subplots(nrows=2,ncols=2,figsize=(10,10)) #sets up a figure object 
#with two rows and two columns of plots, 10x10 "inches"

ur = np.random.rand(10000) #10000 uniform random numbers
nr = np.random.randn(10000) #10000 random numbers drawn according to normal distribution

plt.subplot(2,2,1) #in a 2x2 frame this one is at position 1
plt.hist(ur)
plt.xlabel("Uniform Random Number, Default 10 Bin")

plt.subplot(2,2,2) #this one is at position 2
plt.hist(ur, 20) # display 20 bins
plt.xlabel("Uniform Random Number, 20 Bin")

plt.subplot(2,2,3) #position 3
plt.hist(nr)
plt.xlabel("Normal Random Number, Default 10 Bin")

plt.subplot(2,2,4) #position 4
plt.hist(nr, 20) # display 20 bins
plt.xlabel("Normal Random Number, 20 Bin")

plt.show()
````

````python
#data with noise example
plt.close("all")

N=100
x = np.linspace(0, 2*np.pi, N) #generate N evenly spaced datapoints in range (0,2*pi)
noise = np.random.randn(N) #random bits of noise
signal = 2*np.sin(x)

y = signal + noise
plt.plot(x, signal) # signal + noise
plt.scatter(x, y) #scatter plot

plt.figure()
y = signal + 0.2*noise # singal + 0.2*noise i.e. low noise
plt.plot(x, signal)
plt.scatter(x, y) #scatter plot

plt.figure()
y = signal + 0.2*noise # singal + 0.2*noise i.e. low noise
plt.plot(x, signal)
plt.errorbar(x, y, 0.2*noise) #plot noise as error bars to data instead of scatter plot

plt.show()
````

````python
#pie and bar charts
plt.close("all")

#simple pie chart
x = [30, 20, 15, 25, 10]
dataName = ['data1', 'data2', 'data3', 'data4', 'data5']

plt.figure(figsize=(5,5)) #figsize: output figure window size
plt.pie(x, labels=dataName)

#fancier pie chart
x = [10, 10, 20, 20, 30]
dataName = ['data1', 'data2', 'data3', 'data4', 'data5']
explode = [0.01, 0, 0, 0.04, 0.09] #determines how "big" the separation is at each slice

plt.figure(figsize=(5,5))
# autopct='%.2f %%': %.2f display value upto 2 decimal, 
# %% is used for displaying % at the end
plt.pie(x, explode=explode, labels=dataName, autopct='%.2f%%') 

#simple bar chart
x =[1,2,3,4,5]
heights = np.random.rand(1,len(x))[0] #generate a vector of random numbers the same length as x
plt.figure(figsize=(5,5))
plt.bar(x,heights)

#more advanced bar chart
A = [2, 4, 8, 6]
increment = [5, 6, 4, 1]
years = [2005, 2010, 2015, 2020]

plt.subplot(2,1,1) #1st plot in a 2x1 layout 
plt.bar(years, A, color = 'b', label='A')
plt.bar(years, increment, color = 'r', bottom = A, label='increment') #notice bottom = A
plt.legend()

##### subplot 2
x = [1, 2, 4]
y = [3.5, 3, 2]
z = [2, 3, 1.5]

width = 0.2
locs = np.arange(1, len(x)+1) #locations for bars +1 because arange is not inclusive
plt.subplot(2,1,2) #second plot in our 2x1 layout
plt.bar(locs, x, width=width, label='x') #plot at each location
plt.bar(locs+width, y, width=width, color="red", label='y') #add a new bar a little bit over
plt.bar(locs+2*width, z, width=width, color="black", label='z') #add another bar a little further over
plt.legend()

plt.xticks([1.25, 2.25, 3.25],
          [r'$2000$', r'$2005$', r'$2010$'])

plt.show()
````

````python
#making a plot with annotations! 

# close all the figures, if open from previous commands
plt.close('all') 

############ Sin(x) ####################
x=np.linspace(0, 2*np.pi, 100) 
sinx=np.sin(x) # calculate sin(x)
plt.plot(x,sinx, label='sin') # legend 

############ Legend ####################
plt.legend(loc="best") #show legend

#### Lable and Grid ####################
plt.xlabel("Radian") # x label 
plt.ylabel("Amplitude") # y label

############# Annotate ######################
#1. 
# other arrow style 
# '-', '->', '-[', '<-', '<->', 'fancy', 'simple', 'wedge' 
#sin(pi/2)=1
plt.annotate(r'$sin(\frac{\pi}{2})=1$', 
             fontsize=16,
             xy=(np.pi/2, 1), xytext=(1 , 0.5),
             arrowprops=dict(arrowstyle="->"),
        )

# 2. 
#=============================================================
# width: The width of the arrow in points
# frac: The fraction of the arrow length occupied by the head
# headwidth: The width of the base of the arrow head in points
# shrink: Moves the tip and the base of the arrow some percent 
# away from the annotated point and text, 
#=============================================================
#sin(3*pi/2)=-1
plt.annotate(r'$sin(\frac{3\pi}{2})=-1$', 
             fontsize=18,
             xy=(3*np.pi/2, -1), xytext=(1.5 , -0.5),
             arrowprops=dict(facecolor='red', shrink = 0.04, 
                             connectionstyle="arc3,rad=.2"),
        )       
     
# 3.          
#################### Add text to plot ###########           
plt.text(5, 0.5, 'This is \nSine wave');
        
plt.show() #display the plot
````

````python
#A final fun example of displaying a model solar system
AU = 1.5e11 #m (one 1 AU = average earth sun distance)
thetaSpace = np.linspace(0,2*np.pi,1000) #1000 points between 0 and 2 pi
x = [AU*np.sin(theta) for theta in thetaSpace] #generate x and y values for circle of r = 1 AU
y = [AU*np.cos(theta) for theta in thetaSpace]
rSun = 6.957e8 #m, the radius of the Sun
sun = plt.Circle((0,0),rSun,color='yellow') #make a circular Sun-sized plot object centered at the origin
fig, ax = plt.subplots(figsize=(10,10)) 
ax.add_artist(sun) #adds circle to represent Sun as calculated above
ax.set_facecolor('xkcd:black') #change the background to black
plt.plot(x,y,label="Circular approximation of Earth's orbit")
legendSun = ax.legend([sun],["Sun (to scale)"], loc="upper right") #required to get both sun and earth orbit labels
ax = plt.gca().add_artist(legendSun) #manually adding first legend
plt.legend(loc="lower right") #display second legend with Earth orbit label
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.title("The Earth and Sun"+r"$\approx$"+" to scale") #the r indicates a LaTeX string
plt.gca().set_aspect('equal') #make the axes equal

#adding fake stars
starsX = []
starsY = []
for i in range(100):
    num = random.randint(-1.5e11,1.5e11)
    starsX.append(num)
for i in range(100):
    num = random.randint(-1.5e11,1.5e11)
    starsY.append(num)
for i in range(len(starsX)): 
    plt.plot(starsX[i],starsY[i],'w.',markersize=.5)
    
plt.show()
````

### 3: saving figures

Saving is easy. Once you're done with your plot and it's ready to save, at this command to the bottom of your code: 
```python
plt.savefig('myfile.png')``` 

Where 'myfile' contains the path and name of the image to be saved (by default images are saved in your working directory) and the '.png' is the file extension (other common extensions like .jpeg should work as well). You can also pass optional arguments to `savefig()`, most of which are very specific (and you can read the documenation if you're curious). Two that are useful in more than a handful of scenarios are the `dpi` and `transparent` commands, ie:
```python
plt.savefig('myfile.png',dpi=300,transparent=True)
```
This will generate a high resolution version of your plot with a transparent background.
